```python
def create_ticket_text(row):
    return f"""
Issue:
{row['initial_message']}

Resolution:
{row['resolution_summary']}
"""
```

```python
def expand_ticket_and_fix_None(text, llm):
    prompt = f"""
You are an expert technical support documentation assistant.

Your task is to improve and expand a support ticket while preserving its original meaning.

STRICT RULES:
- Do NOT invent or assume facts that are not present in the ticket.
- Do NOT change the original intent or meaning.
- Keep the content professional, clear, and concise.
- Expand the issue into 2–3 well-written sentences.
- Expand the resolution into 2–3 well-written sentences.
- If the resolution is missing, empty, or "None":
  - Provide a realistic, practical solution based ONLY on the issue described.
  - Do NOT hallucinate specific systems, tools, or environments unless clearly implied.

OUTPUT FORMAT (MUST follow exactly):

Issue:
<expanded issue>

Resolution:
<expanded resolution>

Ticket:
{text}
"""
    response = llm.invoke(prompt)
    return response
```

```python
import pickle
import time
```

```python
expanded_docs = []
```

```python
len(expanded_docs)
```

```python
start_idx = len(expanded_docs)
```

```python
for i in range(start_idx, len(df_unique)):
    row = df_unique.iloc[i]
    ticket_text = create_ticket_text(row)

    retries = 5
    for attempt in range(retries):
        try:
            time.sleep(5)

            expanded = expand_ticket_and_fix_None(ticket_text, llm)
            print(f'{i} Done')

            if expanded is None:
                print(f"Warning: row {i} returned empty response, skipping")
                break

            expanded_docs.append(expanded)

            with open("expanded_docs.pkl", "wb") as f:
                pickle.dump(expanded_docs, f)

            break 

        except Exception as e:
            if "429" in str(e):
                wait_time = (2 ** attempt) * 5
                print(f"Rate limit hit. Waiting {wait_time}s...")
```

```python
len(expanded_docs)
```

```python
documents = []
metadatas = []

for i, expanded in enumerate(expanded_docs):
    row = df_unique.iloc[i] 
    
    doc = f"""
    Category: {row['refined_issue']}
    {expanded}
    """
    metadata = {
        "ticket_id": str(i),
        "category": row["refined_issue"],
    }

    documents.append(doc)
    metadatas.append(metadata)
```

```python
from langchain_core.documents import Document
final_documnets = []
for i in range(len(documents)):
    doc = Document(page_content=documents[i],metadata=metadatas[i])
    final_documnets.append(doc)
```

```python
with open('final_documnets.pkl' ,'wb') as f:
    pickle.dump(final_documnets,f)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import Chroma
vectorDB = Chroma.from_documents(documents=final_documnets,embedding=embedding_model,collection_name="support_tickets",persist_directory='./chromadb')
```

```python
retriever = vectorDB.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 5, "fetch_k": 20}
)
```

```python
from langchain_community.retrievers import BM25Retriever
from langchain_classic.retrievers.ensemble import EnsembleRetriever
```

```python
bm25 = BM25Retriever.from_documents(final_documnets)
bm25.k = 5 
```

```python
hybrid_retriever = EnsembleRetriever(
    retrievers=[bm25, retriever],
    weights=[0.4, 0.6]
)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
def improve_user_query(input_query, llm):
    prompt_template = PromptTemplate(
        input_variables=["input_query"],
        template="""
You are an expert query rewriting assistant for a support ticket retrieval system.

Your goal is to improve the user query for better search results.

STRICT RULES:
- Preserve the original meaning exactly. Do NOT change intent.
- Fix spelling and grammar mistakes.
- Rewrite the query to be clear, concise, and specific.
- Add relevant keywords ONLY if they are strongly implied.
- Keep important technical terms, error codes, and product names unchanged.
- Do NOT add explanations, only return the improved query.
- Return a single  clear query.

OUTPUT:
<improved_query>

User query:
{input_query}
"""
    )

    fixed_query = llm.invoke(prompt_template.format(input_query=input_query))
    return fixed_query.content.strip()
```

```python
prompt_template = """
You are an expert technical support assistant.

Answer the user's question using ONLY the provided context.

STRICT RULES:
- Use ONLY the context. Do NOT hallucinate.
- If the answer is not clearly supported, respond EXACTLY:
  [I can not answer now !]
- Be polite, clear, and actionable.
- Keep the answer concise (1–2 sentences or short steps).
- Use simple Markdown ONLY if helpful:
  - bullet points (-)
  - numbered steps (1. 2. 3.)
- Do NOT over-format.

Context:
{context}

Question:
{question}

Answer:
"""
```

```python
prompt = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)
```

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
chain = prompt | llm | StrOutputParser()
```

```python
user_query = 'i can nt loggin in my acount'
final_query = improve_user_query(user_query,llm)
```

```python
context = hybrid_retriever.invoke(final_query)
```

```python
response = chain.invoke({
    "context":format_docs(context),
    'question' : final_query
})
print(response)
```
